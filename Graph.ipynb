```python
### 라이브러리 불러오기

import pandas as pd
from pandas import DataFrame
import numpy as np
```

```python
### 지역별 파일 불러오기

f = pd.read_csv("서울_구별_확진자_동향.csv",header = None,)
f.transpose()
f.columns = f.iloc[0]
f = f.drop(f.index[0])
f.index = f['지역']
f = f.set_index('지역',drop = True)
f
```

```python
### 서울시 구별 리스트 저장

area = list(f.index)[:-1]
area
index = list(f.index)
index
```

```python
### 구별 누적 확진자 수 데이터 변수 생성

total = []
for i in index:
    total.append(sum(list(map(int,f.loc[i][1:]))))
total
total_list = []
for i,v in zip(area,total):
    total_list.append((i,v))
total_list
```

```python
### 누적 변수 데이터 컬럼 추가 

f['total'] = total
f
```

```python
### 각 구에 따른 월별 확진자 수를 딕셔너리로 저장 

area_dict = {}
for i,v in enumerate(area):
    area_dict[v] = list(map(int,f.loc[v][1:18]))
```

```python
area_dict
```

```python
### 지도그리기에 필요한 모듈 불러오기 

import geopy
from geopy.geocoders import Nominatim
from pprint import pprint
```

```python
lat_long = {}
for key in list(area_dict.keys()):
    service = geopy.Nominatim(user_agent = "myGeocoder")  
    # Nominatim 서비스 객체에 대한 핸들러 가져 오기
    locationObj = service.geocode(key+'청')
    # 각 구청의 위치정보 받아오기
    lat_long[key+'청'] = (locationObj.latitude,locationObj.longitude)
    # 위치정보에서 위도 경도 받아오기
lat_long['강서구청'] = (37.5657617,126.8226561)
# 강서구청이 부산 강서구청으로 인식되어 새롭게 지정
lat_long['강동구청'] = (37.5492077,127.1464824)
# 강동구청이 송파구에 위치해 새롭게 지정
```

```python
### 지도에 표시하기 위한 모듈 불러오기 
import folium

# 각 구별 경계선 정보 
geo_json = 'https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo.json'

# 각 구별 누적 확진자 데이터 프레임으로 저장
data = pd.DataFrame(total_list,columns = ['name','value'])
```

```python
### 서울 특별시 지도 첫 화면 설정 

seoul_map = folium.Map(location = [37.5502, 126.982], zoom_start = 10.5, tiles= 'cartodbpositron')


folium.Choropleth(
    geo_data = geo_json,                  # 자료구조
    name = 'choropleth',                  # 이름설정
    data = data,                          # 데이터 선택
    columns = ['name','value'],           # 구랑 누적데이터 선택
    key_on = 'feature.properties.name',   # json 파일에서 각 구별로 가는 경로
    fill_color = 'YlOrRd',                # 색 지정 
    fill_opacity = 0.7,                   # 구역 투명도 
    line_opacity = 0.5,                   # 경계선 투명도 
).add_to(seoul_map)
```

```python

from folium import IFrame
import base64
for key,name in zip(area,list(lat_long.keys())):
    pic = base64.b64encode(open('img/'+str(key)+'.jpg','rb').read()).decode() # 인코딩하여 이미지 파일 경로변수로 저장
    image_tag = '<img src="data:image/jpeg;base64,{}">'.format(pic) # pic변수를 html형식으로 새로운 변수에 할당
    iframe = folium.IFrame(image_tag, width=500, height=350) # IFrame 생성
    popup = folium.Popup(iframe, max_width=1000)
    icon = folium.Icon(color="red", icon="ok") # 잘 보이는 아이콘 생성
    folium.Marker([lat_long[name][0], lat_long[name][1]], popup=popup,icon = icon).add_to(seoul_map) # 마커에 위치, 기능 추가
seoul_map.save(outfile='Seoul_Corona_map.html') # html형식의 파일로 저장
```
